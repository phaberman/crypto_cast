```python
# Import packages 
import pandas as pd
from prophet import Prophet
```

```python
# Function to import data
def import_data():
    path = 'data/btc_clean.csv'
    return pd.read_csv(path)
```

```python
df = import_data()
```

```python
# Train/test split
df_train = df.iloc[:1789]
df_test = df.iloc[1789:]
```

```python
model = Prophet(seasonality_mode='multiplicative')
model.fit(df_train)
```

```python
forecast_test = model.predict(df_test)
forecast_test['yhat'] - df_test['y']
```

```python
model.plot(forecast_test)
```

```python
# Import packages
import pandas as pd
```

```python
# A function that loads the dataframe
def import_data():
    path = 'data/btc.csv'
    return pd.read_csv(path)
```

```python
# Function to rename the data columns
def rename_columns(df):
    df.columns = ['ds', 'y']
    return df
```

```python
# Function to set datatype of columns
def set_dtypes(df):
    df['ds'] = pd.to_datetime(df['ds'])
    return df
```

```python
# Function that combines previous functions into a pipline
def pipeline():
    df = import_data()
    df = rename_columns(df)
    df = set_dtypes(df)
    return df.to_csv('data/btc_clean.csv')
```

```python
pipeline()
```
